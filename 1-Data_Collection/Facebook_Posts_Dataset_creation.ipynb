# **Now we combine the different dataframes into 1 dataframe that contains all the scraped posts.**

```python
frames = [data_radiomed, data_jawhra, data_nessma,data_musaique,data_shemsfm]
posts_df = pd.concat(frames)
```

```python
posts_df.reset_index(inplace=True)
```

```python
posts_df.to_json('/content/drive/MyDrive/3oumara/news_scraping/all_posts.json')
```
